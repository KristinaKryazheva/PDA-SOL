```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gender_counts = df["Gender"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Распределение по гендеру")
plt.show()
```

Большую долю - 91,5% составляют мужчины

```python
df['Age'].plot.hist(bins=20)
plt.xlabel('Возраст')
plt.ylabel('Количество игроков')
plt.title('Распределение возраста игроков')
plt.show()
```

```python
nationalities = df['Nation'].value_counts().head(10)
nationalities.plot(kind='pie')
plt.title('Топ-10 национальностей игроков')
plt.show()
```

Большую долю игроков занимают англичане и немцы. Также значительную долю имеют испанцы, французы,аргентинцы

```python
df.boxplot(column=['Shooting', 'Passing', 'Dribbling', 'Defending', 'Physicality'])
plt.title('Распределение показателей по типам навыков')
plt.xlabel('Навык')
plt.ylabel('Значение')
plt.show()
```

```python
df_sorted = df.sort_values(by='Strength', ascending=False)
top_10_players = df_sorted.head(10)
plt.bar(top_10_players['Name'], top_10_players['Strength'])
plt.title('Топ 10 игроков по показателю силы')
plt.xlabel('Игрок')
plt.ylabel('Сила')
plt.xticks(rotation=45)

plt.show()
```

Наивысшие показатели силы имеют Daryl Dike и Tomas Chory

```python
average_age = df[df["Nation"] == "England"]["Age"].mean()

print("Средний возраст футболистов из Англии:", round(average_age))
```

```python
penalties_by_nation = df.groupby("Nation")["Penalties"].sum()
penalties_by_nation = penalties_by_nation.sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
penalties_by_nation.plot(kind="bar")
plt.title("Количество забитых пенальти по национальностям")
plt.xlabel("Национальность")
plt.ylabel("Количество пенальти")
plt.show()
```

Англичане забили большее количество пенальти
