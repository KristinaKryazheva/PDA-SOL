```python
import math
import random


no_of_balls = 25
x = [random.triangular() for i in range(no_of_balls)]
y = [random.gauss(0.5, 0.25) for i in range(no_of_balls)]
colors = [random.randint(1, 4) for i in range(no_of_balls)]
areas = [math.pi * random.randint(5, 15)**2 for i in range(no_of_balls)]
plt.figure()
plt.scatter(x, y, s=areas, c=colors, alpha=0.85)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.show()
```

```python
sns.histplot(df["MSDS(unadjusted)"], bins=15, kde=True)
sns.histplot(df["Mean Unadjusted Salary (all occupations)"], bins=15, kde=True)
plt.title("Распределение по средней зарплате разработчиков и средней зарплате всех специвльностей")
plt.legend(['MSDS(unadjusted)','Mean Unadjusted Salary (all occupations)'])
```

```python
# Средня арендная плата по городам
plt.figure(figsize=(12, 6))
plt.plot(df['City'], df['Rent avg'], marker='o')
plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Average Rent')
plt.title('Average Rent by City')
plt.show()
```

```python
#количество вакансий разработчиков по городам
plt.figure(figsize=(12, 6))
plt.bar(df['City'], df['Number of Software Developer Jobs'])
plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Number of Jobs')
plt.title('Number of Software Developer Jobs by City')
plt.show()
```

```python
#топ 20 городов по средней зарплате
sorted_df = df.sort_values('Mean Unadjusted Salary (all occupations)', ascending=False)
top_20_cities = sorted_df.head(20)
plt.figure(figsize=(12, 6))
plt.bar(top_20_cities['City'], top_20_cities['Mean Unadjusted Salary (all occupations)'])
plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Mean Unadjusted Salary (all occupations)')
plt.title('Top 20 Cities by Salary')
plt.show()
```

```python
#топ 20 городов по средней з.п. всех профессий и арендная плата в этих городах
sorted_df = df.sort_values('Mean Unadjusted Salary (all occupations)', ascending=False)
top_20_cities = sorted_df.head(20)

plt.figure(figsize=(12, 6))
plt.bar(top_20_cities['City'], top_20_cities['Mean Unadjusted Salary (all occupations)'], label='Salary')
plt.bar(top_20_cities['City'], top_20_cities['Rent avg'], label='Rent')

plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Amount')
plt.title('Top 20 Cities by Salary and Rent')
plt.legend()
plt.show()
```

```python
sorted_d = df.sort_values('MSDS(adjusted)', ascending=False)
top_20_cities = sorted_d.head(20)

plt.figure(figsize=(12, 6))
plt.bar(top_20_cities['City'], top_20_cities['MSDS(adjusted)'], label='Salary')
plt.bar(top_20_cities['City'], top_20_cities['Cost of Living Plus Rent avg'], label='Living cost')

plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Amount')
plt.title('Top 20 Cities by Salary and Rent')
plt.legend()
plt.show()
```

```python
#Доля аренды в з.п.(все спец.) в New York
ny_data = df[df['City'] == 'New York, NY']

salary = ny_data['Mean Unadjusted Salary (all occupations)']
rent = ny_data['Rent avg']

rent_share = (rent / salary) * 100
labels = ['Зарплата', 'Аренда']
sizes = [100, rent_share]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Доля аренды в зарплате в Нью-Йорке')

plt.show()
```

```python
#3d график
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

salary = df['Mean Unadjusted Salary (all occupations)']
jobs = df['Number of Software Developer Jobs']
home_prices = df['Median Home Price']

ax.scatter(salary, jobs, home_prices, c='r', marker='o')

ax.set_xlabel('Mean Unadjusted Salary')
ax.set_ylabel('Number of Software Developer Jobs')
ax.set_zlabel('Median Home Price')
plt.show()
```
